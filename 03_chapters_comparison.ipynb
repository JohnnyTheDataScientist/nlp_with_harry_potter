## reading chapter names

```python
with open('harry_potter_chapter_names.txt', 'r', encoding="utf8") as f:
    chapters = f.readlines()
```

```python
chapters
```

```python
## remove the special new line character at the end of each chapter
for idx, chapter in enumerate(chapters):
    chapters[idx] = chapters[idx].replace('\n', '')
```

```python
chapters
```

## find out a property of one chapter

```python
## select the chapter of interest
chapter_name = chapters[30]
print(chapter_name)
```

```python
## read the chapter content into the object, "lines"
with open('harry_potter_chapters/' + chapter_name + '.txt', 'r', encoding="utf8") as f:
    lines = f.readlines()

## find out the length of the chapter
assert len(lines) == 1
print('The length of ' + chapter_name + ' is ' + str(len(lines[0])))
```

```python
## find out the number of occurrence of the word "Harry"
word = 'Ron'
print('Number of occurrence of ' + word +':', lines[0].count(word))
```

## find out a property of all chapters

```python
## get the chapter lengths
chapter_len = []

## loop through each chapter
for chapter_name in chapters:

    ## for each chapter, read the chapter content
    with open('harry_potter_chapters/' + chapter_name + '.txt', 'r', encoding="utf8") as f:
        lines = f.readlines()
    assert len(lines) == 1
    
    ## store the property into a list
    chapter_len.append(len(lines[0])) 
```

```python
chapter_len
```

```python
## get the number of occurrence of the word "Harry"
word_occurance = []
word = 'Harry'

## loop through each chapter
for chapter_name in chapters:

    ## for each chapter, read the chapter content
    with open('harry_potter_chapters/' + chapter_name + '.txt', 'r', encoding="utf8") as f:
        lines = f.readlines()
    
    ## store the property into a list
    word_occurance.append(lines[0].count(word)) 
```

```python
word_occurance
```

## the chapter with the least and the largest property values

```python
## find out the longest chapter
max_idx = chapter_len.index(max(chapter_len))
print(chapters[max_idx] + ' is the longest chapter')

## find out the shortest chapter
min_idx = chapter_len.index(min(chapter_len))
print(chapters[min_idx] + ' is the shortest chapter')
```

```python
## find out the chapter in which 'Harry' appears the most
max_idx = word_occurance.index(max(word_occurance))
print('The word "Harry" appears the most in ' + chapters[max_idx])

## find out the chapter in which 'Harry' appears the least
min_idx = word_occurance.index(min(word_occurance))
print('The word "Harry" appears the least in ' + chapters[min_idx])
```

## put the results into a figure

```python
import matplotlib.pyplot as plt

## plot the chapters vs their number of characters
fig = plt.figure(figsize = (10, 40))
plt.margins(y=0)
# creating the bar plot
plt.barh(chapters, chapter_len)

plt.xlabel("Number of Characters")
# plt.ylabel("")
# plt.title("")
plt.show()
```

```python
## save the image into a png file
fig.savefig('analysis_images\chapter_characters_numbers.png', bbox_inches='tight')
```
