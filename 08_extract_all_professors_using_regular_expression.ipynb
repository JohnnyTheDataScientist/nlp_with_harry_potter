### import libaries

```python
import re
import os
```

### read the entire novel series into a long long string

```python
### obtain the list of chapter files
dir_list = os.listdir('./harry_potter_chapters')

### loop through each chapter
long_string = ''
for i in dir_list:
    with open('harry_potter_chapters/' + i, 'r', encoding="utf8") as f:
        chapter_content = f.readlines()[0]

    long_string = long_string + ' '
    long_string = long_string + chapter_content
    
```

### Search for the word "Professor" and its context

```python
pattern = re.compile(r'.{30}(?:Professor|professor|Prof.).{30}')
professor_context = pattern.finditer(long_string)
for i in professor_context:
    print(i.group(0))
```

### write the result into a file 

```python
pattern = re.compile(r'.{30}(?:Professor|professor|Prof.).{30}')
professor_context = pattern.finditer(long_string)
counter = 0

# delete the file if it already exists
try: os.remove('harry_potter_professor_context_1.txt')
except: pass
    
for i in professor_context:
    ### write the contexts to a txt file
    with open('harry_potter_professor_context_1.txt', 'a', encoding="utf8") as f:
        f.write(i.group(0) + '\n')
        counter = counter + 1
```

```python
print(counter)
```

### only write the unique result into a file 

```python
pattern = re.compile(r'.{30}Professor ([A-Z]\w*).{30}')
professor_context = pattern.finditer(long_string)
counter = 0
list_of_professors = []

# delete the file if it already exists
try: os.remove('harry_potter_professor_context_2.txt')
except: pass
        
for i in professor_context:
    if i.group(1) in list_of_professors:
        continue
    else:
        list_of_professors.append(i.group(1))
        ### write the contexts to a txt file    
        with open('harry_potter_professor_context_2.txt', 'a', encoding="utf8") as f:
            f.write(i.group(0) + '\n')
            counter = counter + 1
```

```python
print(counter)
```

```python
print(list_of_professors)
```

## extract proper nouns

```python
%%time
pattern = re.compile(r'(?<![\.\?\!\…\"\”\;\:*])[ ]+((([A-Z][A-Za-z\-.]*[a-z])[,]? )+)')
proper_nouns = pattern.finditer(long_string)
counter = 0
list_of_proper_nouns = []

# delete the file if it already exists
try: os.remove('harry_potter_proper_nouns_1.txt')
except: pass
try: os.remove('harry_potter_proper_nouns_context_1.txt')
except: pass

for i in proper_nouns:
    proper_noun = i.group(1).strip().replace(',', '')
    if proper_noun in list_of_proper_nouns:
        continue
    else:
        list_of_proper_nouns.append(proper_noun)
        ### write the contexts to a txt file
        start = i.span()[0] - 30
        end = i.span()[1] + 30
        if start < 0:
            start = 0
        if end >= len(long_string):
            end = len(long_string) - 1
        with open('harry_potter_proper_nouns_1.txt', 'a', encoding="utf8") as f:
            f.write(proper_noun + '\n')
            counter = counter + 1
        with open('harry_potter_proper_nouns_context_1.txt', 'a', encoding="utf8") as f:
            f.write(long_string[start:end] + '\n')
            counter = counter + 1
```
